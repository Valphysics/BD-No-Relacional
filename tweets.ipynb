```python
#cambiar la estructura de 1000 tweets
db = client['AnalisisdeDatos']
tweets_original = db['Tweets']
cursor = list(tweets_original.find({}, limit=1000))
```

```python
def transformar_tweet(tweet):
    return {
        "usuario": {
            "id": tweet.get("tweet_id"),  # cambia esto si tienes otro campo para ID
            "zona_horaria": tweet.get("user_timezone")
        },
        "contenido": tweet.get("text"),
        "motivo_negativo": tweet.get("negativereason"),
        "sentimiento": tweet.get("airline_sentiment"),
        "aerolinea": tweet.get("airline"),
        "es_retweet": tweet.get("retweet_count", 0) > 0
    }
```

```python
nueva_estructura = [transformar_tweet(doc) for doc in cursor]

nueva_coleccion = db['TweetsEstructurados']
nueva_coleccion.insert_many(nueva_estructura)
from pprint import pprint

for doc in db['TweetsEstructurados'].find().limit(3):
    pprint(doc)

```
